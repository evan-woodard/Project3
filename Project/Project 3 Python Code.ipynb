```python
# Import necessary modules
import sqlalchemy
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect
import json
import pandas as pd
import sqlite3
import numpy as np
```

```python
cnx = sqlite3.connect('Project3')
df = pd.read_sql_query("SELECT * FROM cars", cnx)
df.head()
```

```python
columns = ['County', 'ModelYear', 'Make','Model', 'ElectricVehicleType','ElectricRange']
columned_df = df[columns]
columned_df.head(5)
```

```python
bev_df =columned_df[columned_df['ElectricVehicleType']=='Battery Electric Vehicle (BEV)']
bev_df.head(5)
```

```python
bev_df['ElectricRange'] = bev_df['ElectricRange'].astype(int)
final_df = bev_df[bev_df['ElectricRange']!= 0]
final_df.head(5)
```

```python
bar_df = final_df.groupby(['Model']).mean().reset_index()
bar_df['ElectricRange'] = bar_df['ElectricRange'].round(0).astype(int)
bar_df
```

```python
barData = []
for index, row in bar_df.iterrows():
    bar_plot = {'Model': row['Model'],
              'ElectricRange': row['ElectricRange']}
    barData.append(bar_plot)
    barData = sorted(barData, key=lambda x: x['ElectricRange'], reverse=True)
print(barData)
```

```python
def pieChart(final_df):
    bins = [0, 100, 150, 200, 250, np.inf]
    labels = ['<100', '101-150', '151-200', '201-250', '251+']

    final_df['RangeCategory'] = pd.cut(final_df['ElectricRange'], bins=bins, labels=labels)
    range_counts = final_df['RangeCategory'].value_counts()
    pie_plot = range_counts.to_dict()
    
    return pie_plot
```

```python
pieChart(final_df)
```
